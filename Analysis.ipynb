```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('results.csv')
```

```python
df.head()
```

The following bar chart shows the highest stock price during the first tradinghour of the day by company.

```python
sns.set(style="whitegrid")

hour_df9 = df[(df['hour'] == 9)]

fig = plt.figure(figsize=(15,5))

bar_company = sns.barplot(x="company", y="highest_hourly_price", data=hour_df9)

bar_company.set(xlabel='Company', ylabel='Highest Price at 9AM')

plt.show()
```

The following bar chart shows the highest stock price during the last trading hour of the day by company.

```python
sns.set(style="whitegrid")

hour_df15 = df[(df['hour'] == 12)]

fig = plt.figure(figsize=(15,5))

bar_company = sns.barplot(x="company", y="highest_hourly_price", data=hour_df15)

bar_company.set(xlabel='Company', ylabel='Highest Price at 3PM')

plt.show()
```

The scatter plot matrix below shows the bivariate relationships between a pair of variables, allowing many relationships to be explored in this single chart.

```python
plt.figure(figsize=(15,5))

sns.set(style="ticks")

hue_company = sns.pairplot(df, hue="company")
```

The following line chart compares the hourly highest price between the two stocks, PINS and SNAP. 

```python
pins_df = df[(df['company'] == "pins")]
snap_df = df[(df['company'] == "snap")]

fig = plt.figure(figsize=(15,5))

line_company = sns.lineplot(x="hour", y="highest_hourly_price", data=pins_df, label='pins')
line_company = sns.lineplot(x="hour", y="highest_hourly_price", data=snap_df, label='snap')


line_company.set(xlabel='Hour', ylabel='Price')

plt.show()
```
